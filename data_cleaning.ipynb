```python
#Libraries
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/ek-chris/Practice_datasets/refs/heads/main/EasyVisa%20(1).csv"
def load_data(url):
    """
    Loads and read in data from url

    Params
    input
    url: url 

    output
    df: pd.DataFrame
    """
    df = pd.read_csv(url)
    return df


df = load_data(url)
```

```python
def data_cleaning(data):
    """
    params

    input
    data: pd.DataFrame

    output
    cleaned_data: pd.DataFrame
    """

    #Handle mising values
    missing = data.isna().sum()

    if missing.sum() == 0:
        print("No missing values")
    else:
        data = data.fillna(0)
        print("Number of missing values before cleaning")
        print(missing)
        print("")
        print("No of missing values after cleaning")
        print(data.isna().sum())

    #Handle duplicates
    if data.duplicated().sum() == 0:
        print("No duplicates present in data")
    else:
        print("Number of duplicates before cleaning")
        print(data.duplicated().sum())
        data = data.drop_duplicates()
        print("Number of duplicates after cleaning")


    return data

cleaned_data = data_cleaning(df)
```

```python
def save_data(data):
    data.to_csv("cleaned_data.csv", index = False)
```

```python
save_data(cleaned_data)
```
